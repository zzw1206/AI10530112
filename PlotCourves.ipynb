```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

# datas
x = np.arange(0, 6, 0.1) # 0から6まで0.1刻みで生成みで生成
y1 = np.sin(x)
y2 = np.cos(x)
# plot graphs
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos")
plt.xlabel("x") # x軸
plt.ylabel("y") # y軸
plt.title('sin & cos')
plt.legend()
plt.show()
```
